```python
import triton
import triton.language as tl
```

```python
NUM_E = tl.constexpr(4)


@triton.jit
def add_tensors_list(target, M,N):

    class Items:
        def __init__(self, num_items):
            self.num_items = num_items
            self.items = {}

        def get(self, i):
            return self.items[i]
    items = Items(NUM_E)
    for i in tl.static_range(NUM_E):
        items.items[i] = tl.randn(M,N)

    a = tl.zeros(M,N)
    for e in tl.static_range(NUM_E):
        a = a + items.get(e)
    indices = tl.arange(M*N)
    tl.store(target+indices, a)
    
```

```python
import torch
M = 16
N = 16

tar = torch.empty((M,N))
```

```python
add_tensors_list[(1,)](tar, M,N)
```
